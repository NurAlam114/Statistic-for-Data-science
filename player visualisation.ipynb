```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('sports_dataset.csv')
dataset
```

```python
student_data_dis_auto = {}

for class_name in dataset['Sports'].unique():
    student_data_dis_auto[class_name] = dataset[dataset['Sports'] == class_name][['Height', 'Weight']].values.tolist()
student_data_dis_auto
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(data=dataset, x='Height', y='Weight', hue='Sports', palette='Set2')
```
