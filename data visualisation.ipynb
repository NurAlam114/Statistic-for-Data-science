```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd
```

```python
iris = load_iris()
print(iris)
```

```python
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
print(iris_df)
```

```python
iris_df['species'] = iris.target
iris_df.head(5)
```

```python
iris_df['species'] = [iris.target_names[i] for i in iris.target]
iris_df.tail(10)
```

```python
plt.figure(figsize=(4,2))
sns.heatmap(iris_df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(4,2))
sns.violinplot(x="species", y="petal width (cm)", data=iris_df)
plt.title("Violin Plot of Petal Width by Species")
plt.show()
```

```python
plt.figure(figsize=(4,2))
sns.histplot(data=iris_df, x="petal length (cm)", kde=True, hue="species", bins=20)
plt.title("Histogram of Petal Length")
plt.show()
```

```python
plt.figure(figsize=(4,2))
sns.scatterplot(data=iris_df, x="sepal length (cm)", y="sepal width (cm)", hue="species")
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
plt.figure(figsize=(4,2))
sns.scatterplot(data=iris_df, x="sepal length (cm)", y="sepal width (cm)", hue="species")
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
plt.figure(figsize=(4,2))
sns.lineplot(data=iris_df.drop("species", axis=1))
plt.title("Line Plot of All Features")
plt.show()
```

```python
plt.figure(figsize=(4,2))
sns.barplot(x="species", y="petal width (cm)", data=iris_df, estimator='mean')
plt.title("Average Petal Width per Species")
plt.show()
```

```python
plt.figure(figsize=(4,2))
sns.countplot(x="species", data=iris_df)
plt.title("Count of Each Species")
plt.show()
```

```python
plt.figure(figsize=(4,2))
sns.stripplot(x="species", y="sepal width (cm)", data=iris_df, jitter=True)
plt.title("Strip Plot of Sepal Width by Species")
plt.show()
```

```python

g = sns.FacetGrid(iris_df, col="species")
g.map(sns.histplot, "sepal length (cm)", bins=15, kde=True)
plt.show()
```

```python
plt.figure(figsize=(4,2))
sns.swarmplot(x="species", y="petal length (cm)", data=iris_df)
plt.title("Swarm Plot of Petal Length by Species")
plt.show()
```
